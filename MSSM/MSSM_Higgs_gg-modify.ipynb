```python
pythia_fragment_CUEP8M1 = """
from Configuration.Generator.Pythia8CommonSettings_cfi import *
from Configuration.Generator.MCTunes2017.PythiaCP5Settings_cfi import *
from Configuration.Generator.Pythia8PowhegEmissionVetoSettings_cfi import *
from Configuration.Generator.PSweightsPythia.PythiaPSweightsSettings_cfi import *

generator = cms.EDFilter("Pythia8HadronizerFilter",
                         maxEventsToPrint = cms.untracked.int32(1),
                         pythiaPylistVerbosity = cms.untracked.int32(1),
                         filterEfficiency = cms.untracked.double(1.0),
                         pythiaHepMCVerbosity = cms.untracked.bool(False),
                         comEnergy = cms.double(13000.),
                         PythiaParameters = cms.PSet(
        pythia8CommonSettingsBlock,
        pythia8CP5SettingsBlock,
        pythia8PowhegEmissionVetoSettingsBlock,
        pythia8PSweightsSettingsBlock,
        processParameters = cms.vstring(
            __CHANNEL_DECAY_FRAGMENT__
            ),
        parameterSets = cms.vstring('pythia8CommonSettings',
                                    'pythia8CP5Settings',
                                    'pythia8PowhegEmissionVetoSettings',
                                    'pythia8PSweightsSettings',
                                    'processParameters'
                                    )
        )
                         )
                         
ProductionFilterSequence = cms.Sequence(generator)
"""
```

```python
def fragmentsDictCreator (decay_fr):
    dict = { '2016': pythia_fragment_CUEP8M1.replace('__CHANNEL_DECAY_FRAGMENT__',decay_fr),
             '2016APV': pythia_fragment_CUEP8M1.replace('__CHANNEL_DECAY_FRAGMENT__',decay_fr),
             '2017': pythia_fragment_CUEP8M1.replace('__CHANNEL_DECAY_FRAGMENT__',decay_fr),
             '2018': pythia_fragment_CUEP8M1.replace('__CHANNEL_DECAY_FRAGMENT__',decay_fr)
                    }
    return dict
```

```python
LHEproducer = """
import FWCore.ParameterSet.Config as cms

# link to cards:
# __EXAMPLE__

externalLHEProducer = cms.EDProducer("ExternalLHEProducer",
    args = cms.vstring('__GRIDPACK__'),
    nEvents = cms.untracked.uint32(5000),
    numberOfParameters = cms.uint32(1),
    outputFile = cms.string('cmsgrid_final.lhe'),
    scriptName = cms.FileInPath('GeneratorInterface/LHEInterface/data/run_generic_tarball_cvmfs.sh')
)
"""
```

```python
years = ["2016", "2016APV", "2017", "2018"]
processes = ["H","A","h"]
#mass_point = ["130","150","170","190","200","250","300","350","400","450","500","600","700","800","1000","1200","1500"]
tot_events = 10000
gridpacks_dict = {}
example_dict = {}
dataset_names = {}
```

```python
for year in range(len(years)):
    t_example1 = 'https://github.com/cms-sw/genproductions/tree/master/bin/Powheg/production/2017/13TeV/Higgs/MSSM/ggH_MuMu_NNPDF31_13TeV'
    t_example2 = 'https://github.com/cms-sw/genproductions/tree/master/bin/Powheg/production/2017/13TeV/Higgs/MSSM/ggA_MuMu_NNPDF31_13TeV'
    t_example3 = 'https://github.com/cms-sw/genproductions/tree/master/bin/Powheg/production/2017/13TeV/Higgs/MSSM/ggh_MuMu_NNPDF31_13TeV'
    t_datasetname_year = 'MSSMGluGluTo{process}ToMuMu_MA-{M}_Tanb-{tanb}_TuneCP5_13TeV-powheg-pythia8'
    t_gp_year = "/cvmfs/cms.cern.ch/phys_generator/gridpacks/UL/13TeV/powheg/V2/slc7_amd64_gcc700_CMSSW_10_6_19_MSSMGluGluTo{process}ToMuMu_ma{M}_tb{tanb}/v1/slc7_amd64_gcc700_CMSSW_10_6_19_MSSMGluGluTo{process}ToMuMu_ma{M}_tb{tanb}.tgz"    
    tmp_dataset_dict={}
    tmp_gridpack_dict={}
    tmp_example_dict={}
    dataset_names_year = []
    gp_ggf_year = "gp_ggf_" + years[year]
    gp_ggf_year = []

    for process in range(len(processes)):
        #gp_ggf_year = "gp_ggf_" + years[year]
        dataset_names_year = []
        gp_ggf_year = []

        if processes[process] == "H":
            for M, tanb in params_ggH: 
                d = {"process":processes[process], "M": M, "tanb": tanb}
                tmp_example_dict[processes[process]] = t_example1
                gp_ggf_year.append(t_gp_year.format_map(d))
                dataset_names_year.append(t_datasetname_year.format_map(d))
                tmp_dataset_dict[processes[process]] = dataset_names_year
                tmp_gridpack_dict[processes[process]] = gp_ggf_year
        elif processes[process] == "A":
            for M, tanb in params_ggA: 
                d = {"process":processes[process], "M": M, "tanb": tanb}
                tmp_example_dict[processes[process]] = t_example2
                gp_ggf_year.append(t_gp_year.format_map(d))
                dataset_names_year.append(t_datasetname_year.format_map(d))
                tmp_dataset_dict[processes[process]] = dataset_names_year
                tmp_gridpack_dict[processes[process]] = gp_ggf_year
        elif processes[process] == "h":
            d = {"process":processes[process], "M": 125, "tanb": 30}
            tmp_example_dict[processes[process]] = t_example2
            gp_ggf_year.append(t_gp_year.format_map(d))
            dataset_names_year.append(t_datasetname_year.format_map(d))
            tmp_dataset_dict[processes[process]] = dataset_names_year
            tmp_gridpack_dict[processes[process]] = gp_ggf_year
                
    gridpacks_dict[years[year]] = tmp_gridpack_dict
    dataset_names[years[year]] = tmp_dataset_dict
    example_dict[years[year]] = tmp_example_dict

#print(gridpacks_dict['2016']['bbH'])
#gridpacks_dict = {'VBF_HH' : {'2016': gp_vbf_2016, '2017': gp_vbf_2017, '2018': gp_vbf_2018}}                                                                                                              
#for x in gridpacks_dict:                                                                                                                                                                                   
#    print (x)                                                                                                                                                                                              
#    for y in gridpacks_dict[x]:                                                                                                                                                                            
#        print (y,':', gridpacks_dict[x][y])                 
```

```python
gg_mumu = '''                                                                                                                                                                                           
    'Higgs:useBSM = on',
    'POWHEG:nFinal = 1',
    '35:onMode = off', # turn OFF all H decays
    '35:onIfMatch = 13 -13', # turn ON H->tautau
'''
#'SpaceShower:dipoleRecoil = on' 
```

## .csv Production

```python
#MadgraphVersion = re.compile("V5_2\.[0-9]\.[0-9]")
process_pythia_map = {}
tmp_process_pythia_map = {}
for process in processes:
    tmp_process_pythia_map[process] = fragmentsDictCreator(gg_mumu)
process_pythia_map.update(tmp_process_pythia_map)
```

```python
#for year in range(len(years)):
for year in gridpacks_dict:
    if year == "2016":
        pid = 1812
    elif year == "2016APV":
        pid = 1678
    elif year == "2017":
        pid = 1798
    elif year == "2018":
        pid = 1834
    #print(year)
    for process in processes:
        if process == "H":
            with open('./modify/ggF_'+process+'_'+year+'.csv', 'w') as csvfile:
                csvwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csvwriter.writerow(['prepid', 'Dataset name', 'fragment', 'time'])
                i=0
                for M, tanb in params_ggH:
                    tmp_fragment=""
                    #print(gridpacks_dict[year][i])
                    #print(dataset_names[year][i])
                    #version = MadgraphVersion.search(path).group(0)
                    dataset_name = dataset_names[year][process][i]
                    if year == "2016":
                        prepid = "HIG-RunIISummer20UL16wmLHEGEN-0" + str(pid)
                    elif year == "2016APV":
                        prepid = "HIG-RunIISummer20UL16wmLHEGENAPV-0" + str(pid)
                    elif year == "2017":
                        prepid = "HIG-RunIISummer20UL17wmLHEGEN-0" + str(pid)
                    elif year == "2018":
                        prepid = "HIG-RunIISummer20UL18wmLHEGEN-0" + str(pid)
                    tmp_fragment = LHEproducer.replace('__GRIDPACK__',gridpacks_dict[year][process][i]) + '\n' + process_pythia_map[process][year]
                    final_fragment = tmp_fragment.replace('__EXAMPLE__',example_dict[year][process])
                    time = '3.4'
                    csvwriter.writerow([prepid, dataset_name, final_fragment, time])
                    i=i+1
                    pid = pid+1
                    
        elif process == "A":
            with open('./modify/ggF_'+process+'_'+year+'.csv', 'w') as csvfile:
                csvwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csvwriter.writerow(['prepid', 'Dataset name', 'fragment', 'time'])
                i=0  
                for M, tanb in params_ggA:
                    tmp_fragment=""
                    #print(gridpacks_dict[year][i])
                    #print(dataset_names[year][i])
                    #version = MadgraphVersion.search(path).group(0)
                    dataset_name = dataset_names[year][process][i]
                    if year == "2016":
                        prepid = "HIG-RunIISummer20UL16wmLHEGEN-0" + str(pid)
                    elif year == "2016APV":
                        prepid = "HIG-RunIISummer20UL16wmLHEGENAPV-0" + str(pid)
                    elif year == "2017":
                        prepid = "HIG-RunIISummer20UL17wmLHEGEN-0" + str(pid)
                    elif year == "2018":
                        prepid = "HIG-RunIISummer20UL18wmLHEGEN-0" + str(pid)
                    tmp_fragment = LHEproducer.replace('__GRIDPACK__',gridpacks_dict[year][process][i]) + '\n' + process_pythia_map[process][year]
                    final_fragment = tmp_fragment.replace('__EXAMPLE__',example_dict[year][process])
                    time = '1.7'
                    csvwriter.writerow([prepid, dataset_name, final_fragment, time])
                    i=i+1
                    pid=pid+1
                    
        elif process == "h":
            with open('./modify/ggF_'+'x'+'_'+year+'.csv', 'w') as csvfile:
                csvwriter = csv.writer(csvfile, delimiter=',',
                            quotechar='"', quoting=csv.QUOTE_MINIMAL)
                csvwriter.writerow(['Dataset name','Events', 'fragment','notes','Generator','mcdbid','time','size'])
                i=0                
                tmp_fragment=""
                version="V2"
                dataset_name = dataset_names[year][process][i]
                if year == "2016APV":
                    events = round(tot_events*0.54)
                elif year == "2016":
                    events = round(tot_events*0.46)
                else:
                    events = tot_events
                tmp_fragment = LHEproducer.replace('__GRIDPACK__',gridpacks_dict[year][process][i]) + '\n' + process_pythia_map[process][year]
                final_fragment = tmp_fragment.replace('__EXAMPLE__',example_dict[year][process])
                note = dataset_name.replace('_',' ')
                generators="Powheg_" + version + "  Pythia8"
                mcdb_id = '0'
                time = '2'
                size = '105'
                csvwriter.writerow([dataset_name, events, final_fragment, note, generators, mcdb_id, time, size])
```
